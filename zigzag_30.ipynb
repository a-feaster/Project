```python
# printer/gcode parameters

design_name = 'zigzag_30'
nozzle_temp = 200
bed_temp = 60
print_speed = 60
fan_percent = 100
printer_name='ender_3' # generic / ultimaker2plus / prusa_i3 / ender_3 / cr_10 / bambulab_x1 / toolchanger_T0
```

```python
# design parameters

EW = 0.4 # extrusion width
EH = 0.2 # extrusion height (and layer height)
initial_z = EH*0.6 # initial nozzle position is set to 0.6x the extrusion height to get a bit of 'squish' for good bed adhesion
#layers = 105
layers = 1
```

```python
steps = []
layer = 0
for i in range(layers):
  layer = layer + 1
  steps.append(fc.Point(x=100, y=100, z=initial_z+layer*EH)) #1
  steps.append(fc.Point(x=100.7, y=100, z=initial_z+layer*EH)) #2
  steps.append(fc.Point(x=100, y=99.3, z=initial_z+layer*EH)) #3
  steps.append(fc.Point(x=100, y=97.3, z=initial_z+layer*EH)) #4
  steps.append(fc.Point(x=102.7, y=100, z=initial_z+layer*EH)) #5
  steps.append(fc.Point(x=104.7, y=100, z=initial_z+layer*EH)) #6
  steps.append(fc.Point(x=100, y=95.3, z=initial_z+layer*EH)) #7
  steps.append(fc.Point(x=100, y=93.3, z=initial_z+layer*EH)) #8
  steps.append(fc.Point(x=106.7, y=100, z=initial_z+layer*EH)) #9
  steps.append(fc.Point(x=108.7, y=100, z=initial_z+layer*EH)) #10
  steps.append(fc.Point(x=100, y=91.3, z=initial_z+layer*EH)) #11
  steps.append(fc.Point(x=100, y=89.3, z=initial_z+layer*EH)) #12
  steps.append(fc.Point(x=110.7, y=100, z=initial_z+layer*EH)) #13
  steps.append(fc.Point(x=112.7, y=100, z=initial_z+layer*EH)) #14
  steps.append(fc.Point(x=100, y=87.3, z=initial_z+layer*EH)) #15
  steps.append(fc.Point(x=102, y=87.3, z=initial_z+layer*EH)) #16
  steps.append(fc.Point(x=112.7, y=98, z=initial_z+layer*EH)) #17
  steps.append(fc.Point(x=112.7, y=96, z=initial_z+layer*EH)) #18
  steps.append(fc.Point(x=104, y=87.3, z=initial_z+layer*EH)) #19
  steps.append(fc.Point(x=106, y=87.3, z=initial_z+layer*EH)) #20
  steps.append(fc.Point(x=112.7, y=94, z=initial_z+layer*EH)) #21
  steps.append(fc.Point(x=112.7, y=92, z=initial_z+layer*EH)) #22
  steps.append(fc.Point(x=108, y=87.3, z=initial_z+layer*EH)) #23
  steps.append(fc.Point(x=110, y=87.3, z=initial_z+layer*EH)) #24
  steps.append(fc.Point(x=112.7, y=90, z=initial_z+layer*EH)) #25
  steps.append(fc.Point(x=112.7, y=88, z=initial_z+layer*EH)) #26
  steps.append(fc.Point(x=112, y=87.3, z=initial_z+layer*EH)) #27
  steps.append(fc.Point(x=112.7, y=87.3, z=initial_z+layer*EH)) #28
  layer = layer + 1
  steps.extend(fc.travel_to(fc.Point(x=100, y=87.3,z=initial_z+layer*EH)))
  steps.append(fc.Point(x=100, y=87.3, z=initial_z+layer*EH)) #1
  steps.append(fc.Point(x=100, y=88, z=initial_z+layer*EH)) #2a
  steps.append(fc.Point(x=100.7, y=87.3, z=initial_z+layer*EH)) #3
  steps.extend(fc.travel_to(fc.Point(x=100, y=88, z=initial_z+layer*EH)))
  steps.append(fc.Point(x=100, y=88, z=initial_z+layer*EH)) #2b
  steps.append(fc.Point(x=100, y=90, z=initial_z+layer*EH)) #4
  steps.append(fc.Point(x=102.7, y=87.3, z=initial_z+layer*EH)) #5
  steps.append(fc.Point(x=104.7, y=87.3, z=initial_z+layer*EH)) #6
  steps.append(fc.Point(x=100, y=92, z=initial_z+layer*EH)) #7
  steps.append(fc.Point(x=100, y=94, z=initial_z+layer*EH)) #8
  steps.append(fc.Point(x=106.7, y=87.3, z=initial_z+layer*EH)) #9
  steps.append(fc.Point(x=108.7, y=87.3, z=initial_z+layer*EH)) #10
  steps.append(fc.Point(x=100, y=96, z=initial_z+layer*EH)) #11
  steps.append(fc.Point(x=100, y=98, z=initial_z+layer*EH)) #12
  steps.append(fc.Point(x=110.7, y=87.3, z=initial_z+layer*EH)) #13
  steps.append(fc.Point(x=112.7, y=87.3, z=initial_z+layer*EH)) #14
  steps.append(fc.Point(x=100, y=100, z=initial_z+layer*EH)) #15
  steps.append(fc.Point(x=102, y=100, z=initial_z+layer*EH)) #16
  steps.append(fc.Point(x=112.7, y=89.3, z=initial_z+layer*EH)) #17
  steps.append(fc.Point(x=112.7, y=91.3, z=initial_z+layer*EH)) #18
  steps.append(fc.Point(x=104, y=100, z=initial_z+layer*EH)) #19
  steps.append(fc.Point(x=106, y=100, z=initial_z+layer*EH)) #20
  steps.append(fc.Point(x=112.7, y=93.3, z=initial_z+layer*EH)) #21
  steps.append(fc.Point(x=112.7, y=95.3, z=initial_z+layer*EH)) #22
  steps.append(fc.Point(x=108, y=100, z=initial_z+layer*EH)) #23
  steps.append(fc.Point(x=110, y=100, z=initial_z+layer*EH)) #24
  steps.append(fc.Point(x=112.7, y=97.3, z=initial_z+layer*EH)) #25
  steps.append(fc.Point(x=112.7, y=99.3, z=initial_z+layer*EH)) #26a
  steps.append(fc.Point(x=112, y=100, z=initial_z+layer*EH)) #27
  steps.extend(fc.travel_to(fc.Point(x=112.7, y=99.3, z=initial_z+layer*EH)))
  steps.append(fc.Point(x=112.7, y=99.3, z=initial_z+layer*EH)) #26b
  steps.append(fc.Point(x=112.7, y=100, z=initial_z+layer*EH)) #28
  steps.extend(fc.travel_to(fc.Point(x=100, y=100,z=initial_z+layer*EH+EH)))
```

```python
# generate and save gcode

gcode_controls = fc.GcodeControls(
    printer_name=printer_name,

    initialization_data={
        'primer': 'front_lines_then_y',
        'print_speed': print_speed,
        'nozzle_temp': nozzle_temp,
        'bed_temp': bed_temp,
        'fan_percent': fan_percent,
        'extrusion_width': EW,
        'extrusion_height': EH})
gcode = fc.transform(steps, 'gcode', gcode_controls)
open(f'{design_name}.gcode', 'w').write(gcode)
files.download(f'{design_name}.gcode')
```
